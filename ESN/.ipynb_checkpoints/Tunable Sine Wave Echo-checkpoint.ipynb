```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Functions to update and compute the output

```python
def update_network(leaking_rate,X,y_fb,u,W,Wfb,Win,f):
    print f(Win.dot(u) +   Wfb.dot(y_fb)).shape
    return (1-leaking_rate) * X + leaking_rate * f(Win.dot(u) + W.dot(X) + Wfb.dot(y_fb))

def compute_output(X,Wout):
    return Wout.T.dot(X)
```

# Initialization of the network

```python
N = 100
X = np.random.rand(N) - 0.5
W = np.random.rand(N,N) - 0.5
Wbf = (np.random.rand(N,1) - 0.5)
Win = np.random.rand(N,1) - 0.5
rhoW = np.max(np.abs(np.linalg.eig(W)[0]))
W *= 0.9 / rhoW
```

# Training
+ 100 iterations are executed without recording, to replace the initial conditions
+ Record the answers during 200 iterations 

```python
u = 0.2*np.sin(np.asarray(range(600))*0.01)
y = (0.5*np.sin(np.asarray(range(600)) * u)).reshape(600,1)
_ = plt.plot(range(len(y)),y)
_ = plt.plot(range(len(u)),u)
#_ = plt.plot(range(len(u)),np.asarray(range(600)) * u)
```

```python
X_full = np.ones((200,100))
for i in range(100):
    X = update_network(0.5,X,y[i],u[i],W,Wbf,Win,np.tanh)
print "oooo"
print X.shape
for i in range(200):
    print X.shape
    X_full[i,:] = X
    X = update_network(0.5,X,y[i+100],u[i+100],W,Wbf,Win,np.tanh)
_=plt.plot(pd.DataFrame(X_full))
```

# Use pseudo inverse to compute $W^{out}$

```python
Wout = np.linalg.pinv(X_full).dot(y[:200,:])
_=plt.plot(pd.DataFrame(Wout))
```

# plot objective function and predicted function

```python
ans = []
for i in range(300):
    X = update_network(0.5,X,y[i+300-1],W,Wbf,np.tanh)
    ans.append(compute_output(X,Wout))

_ = plt.plot(range(300),y[300:])
_ = plt.plot(range(300),ans)
```
